## Import DataSet and Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("fraudTest.csv")

```

```python
df.info()
```

## Data Preprocessing

```python
# Convert transaction date n time and date of birth columns type into date type
# Date type can be used for better visualisation

# Convert 'dob' column to datetime
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
# Convert 'dob' column to datetime
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')
```

```python
# Replace spaces with underscores in column names
df.columns = df.columns.str.replace(' ', '_')

# Print the columns to verify the changes
print(df.columns)
```

```python
# Replace gender (possible cases) value with 0 and 1
# 0 represent female and 1 represent male
df['gender'] = df['gender'].str.lower()

# Replace values based on conditions
df['gender'] = df['gender'].apply(lambda x: 1 if x.startswith('m') else (0 if x.startswith('f') else None))
```

```python
df
```

```python
# Delete the Unnamed column from dataset 
unnamed_columns = [col for col in df.columns if col.startswith('Unnamed:')]
df.drop(unnamed_columns, axis=1, inplace=True)
print(df.columns)
print(unnamed_columns)
```

```python
print(df.describe())
print(df.info())
```

```python
# Check the inconsistent value in dataset and replace value with suitable value 
```

```python
if df.isnull().any().any():
    # Handling missing values based on data types
    for col in df.columns:
        if df[col].dtype == 'object':
            # Replace missing values in object columns with mode
            mode_val = df[col].mode()[0]
            df[col].fillna(mode_val, inplace=True)
        elif df[col].dtype in ['int64', 'float64']:
            # Replace missing values in numeric columns with mean
            mean_val = df[col].mean()
            df[col].fillna(mean_val, inplace=True)

    # Verify if there are any remaining missing values
    missing_values_after = df.isnull().sum()
    print("\nMissing values after handling:")
    print(missing_values_after)
else:
    print("No missing values found in the dataset.")            
            
```

```python
df
```

```python
# Convert all string columns to lowercase
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.lower()

# Display the first few rows to verify
#This is for test. is it in lower or not
print(df.head())

```

```python
# Identify duplicate (if any) delete duplicate records
duplicate = df.duplicated().sum()
if duplicate >0:
    df = df.drop_duplicates()
```

### Top 10 States of US with the Most Fraudulent Transactions

```python
# Group by 'state' and sum the 'is_fraud' column to get the count of fraudulent transactions for each state
fraud_by_state = df.groupby('state')['is_fraud'].sum().sort_values(ascending=False)

# Display top 10 states with the most fraudulent transactions

# Print the state with the most fraudulent transactions
print(fraud_by_state.head(10))  

# Visualize the top 10 states with the most fraudulent transactions
plt.figure(figsize=(10, 8))
fraud_by_state.head(10).plot(kind='bar', color='red')
plt.title('Top 10 States of US with the Most Fraudulent Transactions')
plt.xlabel('State')
plt.ylabel('Number of Fraudulent Transactions')
plt.xticks(rotation=0)
plt.show()
```

### Number of Fraudulent Transactions over Time period from June 2020 to Dec 2020 in US states

```python
# Extract month and year from 'trans_date_trans_time'
df['month'] = df['trans_date_trans_time'].dt.to_period('M')

# Group by month and count the number of fraudulent transactions
fraud_by_month = df[df['is_fraud'] == 1].groupby('month').size()

# Convert period to string with month name
fraud_by_month.index = fraud_by_month.index.to_timestamp().strftime('%B')

# Visualize the number of fraudulent transactions per month
plt.figure(figsize=(12, 6))
fraud_by_month.plot(kind='bar', color='blue')
plt.title('Number of Fraudulent Transactions over Time period from June 2020 to Dec 2020')
plt.xlabel('Months')
plt.ylabel('Number of Fraudulent Transactions')
plt.xticks(rotation=0)
plt.show()
```

## Gender Distribution in Fraudulent Transactions

```python
# Filter the DataFrame for fraudulent transactions
fraudulent_transactions = df[df['is_fraud'] == 1]

# Count fraudulent transactions by gender
fraud_gender_counts = fraudulent_transactions['gender'].replace({0: 'female', 1: 'male'}).value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(fraud_gender_counts, labels=fraud_gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Distribution of Fraudulent Transactions by Gender over Time period from June 2020 to Dec 2020 in US states')
plt.legend(title='Gender', labels=['Female', 'Male'], loc='upper right')
plt.show()
```

```python
# Convert the 'trans_date_trans_time' column to datetime
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])

# Filter fraudulent transactions
fraudulent_transactions = df[df['is_fraud'] == 1].copy()

# Extract the month
fraudulent_transactions['month'] = fraudulent_transactions['trans_date_trans_time'].dt.month

# Group by the month and sum the amounts
monthly_fraud_amount = fraudulent_transactions.groupby('month')['amt'].sum()

# Map month numbers to month names
monthly_fraud_amount.index = monthly_fraud_amount.index.map(lambda x: pd.Timestamp(f'2020-{x:02d}-01').strftime('%B'))

# Plotting the bar chart
plt.figure(figsize=(12, 6))
monthly_fraud_amount.plot(kind='bar', color='skyblue')
plt.title('Monthly Fraud Amount')
plt.xlabel('Month')
plt.ylabel('Total Fraud Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the DataFrame to include only fraudulent transactions
fraud_df = df[df['is_fraud'] == 1].copy()

# Map gender values to labels
fraud_df.loc[:, 'gender'] = fraud_df['gender'].map({0: 'Female', 1: 'Male'})

# Create the count plot grouped by category and gender
plt.figure(figsize=(14, 7))
sns.countplot(data=fraud_df, x='category', hue='gender')
plt.title('Count of Fraudulent Transactions by Category and Gender (June 2020 to Dec 2020 in US states)')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
fraud_df = df[df['is_fraud'] == 1].copy()
# Map gender values to labels
fraud_df.loc[:, 'gender'] = fraud_df['gender'].map({0: 'Female', 1: 'Male'})

# Group by state and gender, then sum the amounts
fraud_amount_by_state_gender = fraud_df.groupby(['state', 'gender'])['amt'].sum().reset_index()

# Group by state to get total fraud amount per state and select the top 5 states
top5_states = fraud_amount_by_state_gender.groupby('state')['amt'].sum().sort_values(ascending=False).head(5).index

# Filter the dataset to include only the top 5 states
top5_fraud_df = fraud_amount_by_state_gender[fraud_amount_by_state_gender['state'].isin(top5_states)]

# Sort the filtered dataset by total fraud amount in descending order
top5_fraud_df = top5_fraud_df.sort_values(by='amt', ascending=False)

# Create the bar plot with gender as a hue
plt.figure(figsize=(14, 7))
sns.barplot(data=top5_fraud_df, x='state', y='amt', hue='gender', order=top5_states)
plt.title('Total Amount of Fraudulent Transactions by State and Gender (Top 5 States, June 2020 to Dec 2020 in US)')
plt.xticks(rotation=45)
plt.xlabel('State')
plt.ylabel('Total Fraud Amount')
plt.legend(title='Gender')
plt.show()
```

```python
# Filter fraudulent transactions
fraudulent_transactions = df[df['is_fraud'] == 1].copy()

# Extract the month and category
fraudulent_transactions['month'] = fraudulent_transactions['trans_date_trans_time'].dt.month

# Create a pivot table for heatmap
heatmap_data = fraudulent_transactions.pivot_table(values='amt', index='category', columns='month', aggfunc='sum')

# Map month numbers to month names
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
               7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
heatmap_data.columns = heatmap_data.columns.map(month_names)

# Plot the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap='Blues')
plt.title('Heatmap of Fraudulent Transactions by Category and Month over Time Period June to December 2020 ')
plt.xlabel('Month')
plt.ylabel('Category')
plt.show()
```

```python
# Extract the month
fraudulent_transactions['month'] = fraudulent_transactions['trans_date_trans_time'].dt.month

# Group by month and gender
monthly_fraud_amount_by_gender = fraudulent_transactions.groupby(['month', 'gender'])['amt'].sum().unstack()

# Map month numbers to month names
monthly_fraud_amount_by_gender.index = monthly_fraud_amount_by_gender.index.map(month_names)

# Plot the line plot
plt.figure(figsize=(14, 8))
monthly_fraud_amount_by_gender.plot(kind='line', marker='o')
plt.title('Monthly Fraudulent Transaction Amounts by Gender')
plt.xlabel('Month')
plt.ylabel('Total Fraud Amount')
plt.legend(title='Gender')
plt.show()
```
